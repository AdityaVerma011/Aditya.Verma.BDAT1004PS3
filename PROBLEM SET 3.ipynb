Occupations
Step 1. Import the necessary libraries
Step 2. Import the dataset from this address.
Step 3. Assign it to a variable called users
Step 4. Discover what is the mean age per occupation
Step 5. Discover the Male ratio per occupation and sort it from the most to the least
Step 6. For each occupation, calculate the minimum and maximum ages
Step 7. For each combination of occupation and sex, calculate the mean age
Step 8. For each occupation present the percentage of women and men


```python
# Step 1: Import necessary libraries
import pandas as pd
import numpy as np

# --- Breakpoint 1 ---

# Step 2: Import the dataset
users_data = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', delimiter='|')

# --- Breakpoint 2 ---

# Step 3: Assign it to a variable called users
users = users_data

# --- Breakpoint 3 ---

# Step 4: Display the mean age per occupation
mean_age_per_occupation = users.groupby('occupation')['age'].mean()
display(mean_age_per_occupation)

# --- Breakpoint 4 ---

# Step 5: Discover the Male ratio per occupation and sort it from the most to the least
male_ratio = users[users['gender'] == 'M'].groupby('occupation').size() / users.groupby('occupation').size()
male_ratio = male_ratio.sort_values(ascending=False)
display(male_ratio)

# --- Breakpoint 5 ---

# Step 6: Calculate the minimum and maximum ages for each occupation
min_max_age_per_occupation = users.groupby('occupation')['age'].agg(['min', 'max'])
display(min_max_age_per_occupation)

# --- Breakpoint 6 ---

# Step 7: For each combination of occupation and sex, calculate the mean age
mean_age_by_gender_and_occupation = users.groupby(['occupation', 'gender'])['age'].mean()
display(mean_age_by_gender_and_occupation)

# --- Breakpoint 7 ---

# Step 8: For each occupation present the percentage of women and men
gender_counts = users.groupby(['occupation', 'gender']).size()
occupation_totals = users.groupby('occupation').size()
women_percentage = gender_counts.loc[:, 'F'] / occupation_totals * 100
men_percentage = gender_counts.loc[:, 'M'] / occupation_totals * 100
gender_percentages = pd.DataFrame({'Percentage Women': women_percentage, 'Percentage Men': men_percentage})
display(gender_percentages)
```

The first three columns are year, month, and day. The remaining 12 columns are
average windspeeds in knots at 12 locations in Ireland on that day.
Step 1. Import the necessary libraries
Step 2. Import the dataset from the attached file wind.txt
Step 3. Assign it to a variable called data and replace the first 3 columns by a proper
datetime index.
Step 4. Year 2061? Do we really have data from this year? Create a function to fix it
and apply it.
Step 5. Set the right dates as the index. Pay attention at the data type, it should be

Step 6. Compute how many values are missing for each location over the entire
record.They should be ignored in all calculations below.

A single number for the entire dataset.
Step 9. Create a DataFrame called loc_stats and calculate the min, max and mean

A different set of numbers for each location.
Step 10. Create a DataFrame called day_stats and calculate the min, max and mean

A different set of numbers for each day.

Treat January 1961 and January 1962 both as January.
Step 12. Downsample the record to a yearly frequency for each location.
Step 13. Downsample the record to a monthly frequency for each location.
Step 14. Downsample the record to a weekly frequency for each location.
Step 15. Calculate the min, max and mean windspeeds and standard deviations of the
windspeeds across all locations for each week (assume that the first week starts on
January 2 1961) for the first 52 weeks.

```python
import pandas as pd
import numpy as np

# Import the dataset from the provided file
data = pd.read_csv('wind.txt', delim_whitespace=True, na_values='NaN')

# Replace the first 3 columns by a proper datetime index
data['Date'] = pd.to_datetime(data[['Yr', 'Mo', 'Dy']].astype(str).agg('-'.join, axis=1))
data = data.set_index('Date')
data = data.drop(['Yr', 'Mo', 'Dy'], axis=1)
print("Data after setting date as index:\n", data.head())

# Function to fix the year and apply it
def fix_year(x):
    if x.year > 2000:
        return x - pd.DateOffset(years=100)
    return x

data.index = data.index.map(fix_year)
print("Data after fixing the year in the date:\n", data.head())

# Set the right dates as the index with the appropriate data type
data.index = pd.to_datetime(data.index)
data.index = data.index.astype('datetime64[ns]')
print("Data index after setting the right data type:\n", data.index)

# Count missing values for each location
missing_per_location = data.isnull().sum()
print("Missing values for each location:\n", missing_per_location)

# Count non-missing values in total
total_non_missing = data.notnull().sum().sum()
print("Total non-missing values:\n", total_non_missing)

# Calculate the mean windspeeds of all locations and times
mean_windspeed = data.mean().mean()
print("Mean windspeed across all locations and times:\n", mean_windspeed)

# Calculate stats for each location
loc_stats = pd.DataFrame({
    'Min': data.min(),
    'Max': data.max(),
    'Mean': data.mean(),
    'Std': data.std()
})
print("Statistics for each location:\n", loc_stats)

# Calculate stats for each day
day_stats = pd.DataFrame({
    'Min': data.min(axis=1),
    'Max': data.max(axis=1),
    'Mean': data.mean(axis=1),
    'Std': data.std(axis=1)
})
print("Daily statistics:\n", day_stats)

# Average windspeed in January for each location
january_avg = data[data.index.month == 1].groupby(data[data.index.month == 1].index.to_period('Y')).mean()
print("Average windspeed in January for each location:\n", january_avg)

# Downsample to yearly frequency for each location
yearly = data.resample('Y').mean()
print("Yearly average windspeed for each location:\n", yearly)

# Downsample to monthly frequency for each location
monthly = data.resample('M').mean()
print("Monthly average windspeed for each location:\n", monthly)

# Downsample to weekly frequency for each location
weekly = data.resample('W').mean()
print("Weekly average windspeed for each location:\n", weekly)

# Weekly statistics for the first 52 weeks
weekly_stats = pd.DataFrame()
weekly_stats['min'] = data.resample('W').min().iloc[:52].min(axis=1)
weekly_stats['max'] = data.resample('W').max().iloc[:52].max(axis=1)
weekly_stats['mean'] = data.resample('W').mean().iloc[:52].mean(axis=1)
weekly_stats['std'] = data.resample('W').std().iloc[:52].std(axis=1)
print("Weekly statistics for the first 52 weeks:\n", weekly_stats)
```

# Question 6
Create a line plot showing the number of marriages and divorces per capita in the
U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset into a DataFrame
marriage_divorce_data = pd.read_csv('us-marriages-divorces-1867-2014.csv')

# Extract the necessary columns for plotting
data_years = marriage_divorce_data['Year']
marriages_per_capita = marriage_divorce_data['Marriages_per_1000']
divorces_per_capita = marriage_divorce_data['Divorces_per_1000']

# Set the size of the figure for the line plot
plt.figure(figsize=(10, 6))

# Create a line plot for the number of marriages per capita
plt.plot(data_years, marriages_per_capita, label='Marriages per 1000', color='blue')

# Create a line plot for the number of divorces per capita
plt.plot(data_years, divorces_per_capita, label='Divorces per 1000', color='red')

# Set the title of the line plot
plt.title('Number of Marriages and Divorces per Capita in the U.S. (1867 - 2014)')

# Label the x-axis as 'Year'
plt.xlabel('Year')

# Label the y-axis as 'Count per 1000'
plt.ylabel('Count per 1000')

# Add a legend to the line plot
plt.legend()

# Enable the grid for the line plot
plt.grid(True)

# Display the line plot
plt.show()
```

# Question 8
Create a horizontal bar chart that compares the deadliest actors in Hollywood. Sort
the actors by their kill count and label each bar with the corresponding actor's name.
Don't forget to label your axes!

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset into a DataFrame
actor_data = pd.read_csv('actor_kill_counts.csv')

# Sort the DataFrame by the 'Count' column in ascending order
sorted_actor_data = actor_data.sort_values(by='Count', ascending=True)

# Set the size of the figure for the bar chart
plt.figure(figsize=(10, 6))

# Create a horizontal bar chart, with the actors' names on the y-axis and their kill counts on the x-axis
plt.barh(sorted_actor_data['Actor'], sorted_actor_data['Count'], color='skyblue')

# Label the x-axis as 'Kill Count'
plt.xlabel('Kill Count')

# Set the title of the bar chart
plt.title('Comparison of Kill Counts of Hollywood Actors')

# Invert the y-axis so that the actor with the highest kill count is at the top
plt.gca().invert_yaxis()

# Label each bar with the corresponding kill count
for index, value in enumerate(sorted_actor_data['Count']):
    plt.text(value, index, str(value), va='center')

# Display the bar chart
plt.show()
```

# Question 9

Create a pie chart showing the fraction of all Roman Emperors that were
assassinated.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset into a DataFrame
emperor_df = pd.read_csv('roman-emperor-reigns.csv')

# Count the number of 'Assassinated' occurrences in the 'Cause_of_Death' column
num_assassinated = emperor_df['Cause_of_Death'].value_counts()['Assassinated']

# Count the total number of Roman emperors
total_num_emperors = len(emperor_df)

# Calculate the percentage of Roman emperors who were assassinated
percentage_assassinated = (num_assassinated / total_num_emperors) * 100

# Calculate the percentage of Roman emperors who died from causes other than assassination
percentage_other_causes = 100 - percentage_assassinated

# Create a pie chart to visualize the data
labels = ['Assassinated', 'Died from Other Causes']
sizes = [percentage_assassinated, percentage_other_causes]
colors = ['lightcoral', 'lightskyblue']
explode = (0.1, 0)  # Separate the 'Assassinated' slice for emphasis

# Set the size of the figure for the pie chart
plt.figure(figsize=(8, 8))

# Create the pie chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

# Ensure the pie chart is drawn as a circle
plt.axis('equal')

# Set the title of the pie chart
plt.title('Percentage of Roman Emperors Assassinated vs Other Causes of Death')

# Display the pie chart
plt.show()
```

# Question 10

Create a scatter plot showing the relationship between the total revenue earned by
arcades and the number of Computer Science PhDs awarded in the U.S. between
2000 and 2009.
